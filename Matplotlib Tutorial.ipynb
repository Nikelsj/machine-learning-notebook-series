### Matplotlib Tutorial

As we need to evaluate my trained model's generalization error or training step info or find the info from the data we have, we need to plot the info for vivualization, as the famous sentence: one graph means thousands words. 

As a module will contain so many contents and methods that we could use, for here I just make some useful and common method and logic we could use. 

Here we go.

```python
# first import the modules
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
```

```python
# we could change the default plot style just like the iphone style could be changed.
# in fact, you could not use this style during your first time plot, and you are used to the default style, 
# then you could change the style with this code
style.use('ggplot')   

# there are just a few styles that we could use
# here are some common style I use during the work: ggplot, fivethirtyeight and seaborn
# in fact, you could use some else style
print("what style we could use: ", plt.style.available)
```

```python
# first we should be familiar with the curve plot
# data: 1 to 10
data = np.arange(1, 11)  

data
```

```python
# the curve plot without any configuration
# you could just use the plot.plot(x, y), x means the x-axis value, y means the y-axis value;
# Noted: you could just change the x-axis value and y-axis value with you want.
plt.plot(range(len(data)), data)

# show the graph, you could see the line with x-axis from 0 to 9 and y-axis from 1 to 10
plt.show()
```

```python
# you could even just plot the data without setting the x-axis, default x-axis will be range(len(data))
plt.plot(data)
```

```python
# here I just change the data with exp function
def ex_func(x):
    return np.exp(x)

x = np.arange(-5, 5, 1)
data = ex_func(x)
```

```python
# after we could plot the curve, then we could set some useful configuration.
plt.plot(x, data)    

# set the title of the plot
plt.title('line plot')
# set the x-axis label name
plt.xlabel('x-axis')
# set the y-axis label name
plt.ylabel('y-axis')

# show 
plt.show()
```

```python
# in fact, we could plot two or more lines in one graph
data_one = np.random.randn(10, )
data_two = np.random.randn(10, )

plt.plot(data_one, label='data one')  # we could set this line label with name, so we could seperate with other lines
plt.plot(data_two, label='data two')

# in fact, 

plt.title('plot with two line')
plt.xlabel('x axis')
plt.ylabel('y axis')

# to seperate these two lines, we have to call the plt.legend() function, if not, we couldn't figure out which is which
# we could 
plt.legend()

plt.show()
```

```python
# we could also plot the two lines with two parts in one graph, we could even with one row and two columns or two rows one column

# fig is the returned figure, ax is what we could plot on it
# subplots(1, 2, figsize=(10, 8)): 1 means 1 row, 2 means 2 columns; we could set the figure size with what we want.
fig, ax = plt.subplots(1, 2, figsize=(10, 6))

# as there are two parts, we use ax[0] to represent the first plot.
ax[0].plot(data_one)
ax[0].set_title('plot one')   # here we have to use the set_*** function
ax[0].set_xlabel('x axis')
ax[0].set_ylabel('y axis')

# plot the second graph, we could set the color of the line: 'b' means 'blue'
ax[1].plot(data_two, color='b')
ax[1].set_title('plot two')
ax[1].set_xlabel('x axis')
ax[1].set_ylabel('y axis')

plt.show()
```

```python
# we could not only plot the curve, we could also plot the point of the dataset
x = np.arange(1, 11)
y = np.random.randn(10, )

# use the plt.scatter to plot the point of the dataset
plt.scatter(x, y)

plt.title('scatter point')
plt.xlabel('x axis')
plt.ylabel('y axis')

# in fact, we could also change x axis range and y axis range by what we want, use plt.axis
# you could see that for the point that are just within the range as we want: x from 0 to 12, y from -3 to 3
plt.axis([0, 12, -3, 3])

plt.show()
```

```python
# after you have learned the basic use case for matplotlib for ploting the curve and and point, 
# we really could do much more, one simple example is that we could use the matplotlib to estimate the distribution 
# of dataset, you may wonder what is distribution and histgram, don't worry, I will explain for you if I have time,
# but you could see first time for what is distribution and histgram

# the common distribution is Gaussian distribution（高斯分布） no matter for machine learning or real world, 
# gaussian distribution is used most! you have to be famaliar with it.
# the gaussian distribution is fixed with mu(均值) and sigma（方差）
# I will explain for you if I have any time.
mu = 0
sigma = 1
x = mu + sigma * np.random.randn(10000, )

# plot the histgram
plt.hist(x, bins=50, normed=1)

plt.show()
```

```python
# in fact, when we want to explore the dataset, I usually don't use the matplotlib, I will use more powfull module 
# called seaborn, but you have to know that for most usecase is based on the matplotlib, so even if you use the seaborn
# module, you have to `import matplotlib.pyplot as plt` first! 
# seaborn will be easier for getting insight for data! I will show you!
import seaborn as sns

# plot the distribution and histgram with just one line
# compare with matplotlib, it's more beautiful
sns.distplot(x, bins=50)

plt.show()
```
