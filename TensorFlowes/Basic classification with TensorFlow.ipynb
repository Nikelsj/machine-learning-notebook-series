```python
# import modules
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np

tf.__version__
```

Here we will use FANISH MNIST dataset. This dataset is really like MNIST digits hand-writen dataset which is image from 0 to 9. For Fashin MNIST, this dataset is just clothes names.

```python
fashion_mnist = keras.datasets.fashion_mnist

# load data to train, test 
(train_data, train_label), (test_data, test_label) = fashion_mnist.load_data()
```

```python
# here is just 10 classes name
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# how much datasets
# as you could see that we will have 60,000 samples to train and 10,000 samples to test
print("Train data shape: ", train_data.shape)
print("Test data shape:", test_data.shape)
```

```python
# Get the unique label
print("Unique label: ", np.unique(train_label))
```

```python
# here I just pick one sample to show what the data is?
# this is T-shirt, right?
plt.imshow(train_data[10])
plt.grid(False)
plt.colorbar()
plt.title(class_names[train_label[10]])
plt.show()
```

```python
# in fact, we should normalize our data to be 0 to 1, as for now, each pixel is from 0 to 255
print("Bigest value of one sample: ", np.max(train_data[0]))

# normalize the data to 0 to 1, really easy just divide current data with 255
xtrain = train_data / 255.
xtest = test_data / 255.

print("Bigest value of one sample: ", np.max(xtrain[0]))
```

```python
# just to example that what data we have for now
# the data is easy to understand right? but with many different kinds with same class.
plt.figure(figsize=(12, 12))

for i in range(16):
    plt.subplot(4, 4, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(xtrain[i], cmap=plt.cm.binary)
    plt.title(class_names[train_label[i]])

plt.show()
```

```python
# then we should build our model, as we have already get our data
# with TensorFlow high level API to build model is really easy!
# I will write one tutorial for Keras in fact, but for now, I just show you how to use it
model = keras.models.Sequential()

model.add(keras.layers.Flatten(input_shape=(28, 28)))
model.add(keras.layers.Dense(128, activation=tf.nn.relu))
model.add(keras.layers.Dense(10, activation= tf.nn.softmax))
```

```python
# then we could get our model structure... Here we just flatten our dataset, 
# then we one hidden layer with 128 units, and with output layer with 10 
# the reason that we use softmax as output is to make out logits to probability
# for each class...
# for keras, we do could get more information, like we will have 101,770 units for our model!
# and we have 3 layers, first is input layer with output with 784 units that's for flatten our dataset with 28*28
# and with hidden layer we have 128 units, with output layer with 10 units represent each class
model.summary()
```

```python
# we have get our model and we have the dataset, so now to train our model?
# but one thing that should remember, do we set the optimizer to change model weights?
# and do we set the loss function that we should evuate our training loss to get the forward pass loss?
# not really right? so here we should compile our model for Keras
# but with compile, we could set our optimizer, loss_function and metrics for each epoch
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
# as we have compile our model, the last step is to fit our model
his = model.fit(xtrain, train_label, epochs=30, batch_size=128, validation_data=(xtest, test_label), verbose=1)
```

You could see that with keras to train model, the whole process includes the training accuracy, training loss, test accuracy and test loss, also with training status for each epoch. I would like to say that **Keras** in **TensorFlow** is really should be learned! Easy but powerful!

```python
# then we could plot the training step accuracy and loss curve.
fig, ax = plt.subplots(1, 2, figsize=(12, 12))

ax[0].plot(his.history['acc'], label='train accuracy')
ax[0].plot(his.history['val_acc'], label='test accuracy')
ax[0].set_title('Accuracy curve')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Scores')

ax[1].plot(his.history['loss'], label='train loss')
ax[1].plot(his.history['val_loss'], label='test loss')
ax[1].set_title('Loss curve')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Loss')

plt.legend()

plt.show()
```

Here I just make a test model that I flatten the data to be dense, and with just one hidden layer, in fact there are some more advance model that we could use to make the model to be more expressible. 

From the training and testing accuracy, loss curve, we could find our model is overfitting to the data, as when we train our model more times with each epoch, the train accuracy is up and up, but with test accuracy is lower than the train accuracy. That's overfitting!

```python
# we could evaluate our model on test data
# you could see that we could just get 89% accuracy on test data, not very high, but also great!
score = model.evaluate(xtest, test_label, batch_size=512)
print("score object:", score)
print("evaluatation loss: ", score[0])
print("evaluation accuracy: %.2f %s" % (score[1] * 100, '%'))
```

```python
# we could also predict with our test data, 
# but one thing to notice is that the prediction is just score
prob = model.predict(xtest)

print("Get 0 probability: ", prob[0])

# but with we want to get the prediction of data, we should get the max index as prediction classes for each row
pred = np.argmax(prob, axis=1)

print("Get first 10 samples predictions:", pred[:10])
```

```python
# Here I just want to show you the correct prediction and incorrect prediction of our model
# with correct label with green, incorrect label with red!
n_selected = 4

fig = plt.figure(figsize=(12, 12))

for  i in range(n_selected * n_selected):
    plt.subplot(n_selected, n_selected, i + 1, )
    plt.imshow(xtest[i].reshape(28, 28))
    if pred[i] == test_label[i]:
        color = 'g'
        title_obj = plt.title(class_names[test_label[i]])
    else:
        color = 'r'
        title_obj = plt.title('correct:%s, pred: %s' % (class_names[test_label[i]], class_names[pred[i]]))
    plt.setp(title_obj, color=color)

plt.show()
```
